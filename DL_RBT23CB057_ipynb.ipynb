single layer

```python
import random

# Step 1: Training data (AND gate)
# Inputs and expected outputs
X = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]
Y = [0, 0, 0, 1]
```

```python
# Step 2: Initialize weights and bias
weights = [random.random(), random.random()]
bias = random.random()
learning_rate = 0.1
epochs = 10
```

```python
# Step 3: Activation function (Step function)
def activation(net):
    if net >= 0:
        return 1
    else:
        return 0
```

```python
# Step 4: Training the perceptron
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        net = X[i][0]*weights[0] + X[i][1]*weights[1] + bias
        output = activation(net)
        error = Y[i] - output

        # Weight update rule
        weights[0] += learning_rate * error * X[i][0]
        weights[1] += learning_rate * error * X[i][1]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Output: {output}, Error: {error}")
```

```python
# Step 5: Testing
print("\nTesting Perceptron:")
for i in range(len(X)):
    net = X[i][0]*weights[0] + X[i][1]*weights[1] + bias
    print(f"Input: {X[i]} -> Output: {activation(net)}")
```

multilayer


```python
import math
import random
```

```python
# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
# Derivative of sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# Training data for XOR
X = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]
Y = [[0], [1], [1], [0]]
```

```python
# Initialize weights randomly
w_ih = [[random.random(), random.random()],
        [random.random(), random.random()]]   # Input → Hidden

w_ho = [[random.random()],
        [random.random()]]
```

```python
# Initialize biases
b_h = [random.random(), random.random()]
b_o = random.random()
learning_rate = 0.5
epochs = 5000
```

```python
# Training using backpropagation
for epoch in range(epochs):
    for i in range(len(X)):

        # ---------- Forward Pass ----------
        h1 = sigmoid(X[i][0]*w_ih[0][0] + X[i][1]*w_ih[1][0] + b_h[0])
        h2 = sigmoid(X[i][0]*w_ih[0][1] + X[i][1]*w_ih[1][1] + b_h[1])

        output = sigmoid(h1*w_ho[0][0] + h2*w_ho[1][0] + b_o)

        # ---------- Backpropagation ----------
        error = Y[i][0] - output
        d_output = error * sigmoid_derivative(output)

        # Hidden layer error
        d_h1 = d_output * w_ho[0][0] * sigmoid_derivative(h1)
        d_h2 = d_output * w_ho[1][0] * sigmoid_derivative(h2)

        # ---------- Update Weights ----------
        w_ho[0][0] += learning_rate * d_output * h1
        w_ho[1][0] += learning_rate * d_output * h2
        b_o += learning_rate * d_output

        w_ih[0][0] += learning_rate * d_h1 * X[i][0]
        w_ih[1][0] += learning_rate * d_h1 * X[i][1]
        w_ih[0][1] += learning_rate * d_h2 * X[i][0]
        w_ih[1][1] += learning_rate * d_h2 * X[i][1]

        b_h[0] += learning_rate * d_h1
        b_h[1] += learning_rate * d_h2
```

```python
# Testing the trained MLP
print("\nXOR Output:")
for i in range(len(X)):
    h1 = sigmoid(X[i][0]*w_ih[0][0] + X[i][1]*w_ih[1][0] + b_h[0])
    h2 = sigmoid(X[i][0]*w_ih[0][1] + X[i][1]*w_ih[1][1] + b_h[1])
    output = sigmoid(h1*w_ho[0][0] + h2*w_ho[1][0] + b_o)

    print(f"Input: {X[i]} -> Output: {round(output)}")
```
